```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(io=r'C:\Users\data_to_cloud\heteropodatoxin.xlsx',index=False)
```

```python
#This code expects an excel sheet with columns to be plotted as boxplot. Put Column names you want in the boxplot.
```

```python
# Sample structure of the dataframe, 
df
```

```python
# Make the box plot

fig, ax = plt.subplots(figsize=(20, 10))
meanpointprops = dict(marker='D', markeredgecolor='black',
                      markerfacecolor='firebrick',markersize=10)
bp = sns.boxplot(data=df,boxprops=dict(facecolor=(0,0,0,0)),showmeans='True',meanprops=meanpointprops)
ax = sns.stripplot(data=df, color=".3",jitter=False,size=10)

# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.ylabel("Amplitude (mV)")
ax.yaxis.label.set_size(15)  # Image size
ax.set_ylim([0,1.2])     # Set y axis limit
ax.tick_params(direction='out')  # Direction of ticks

```

```python
#Use this if you have multiple plots and want specifies colors for each of them.


colors=['red','black','green','blue','purple','maroon'] #Put your in the colors in the list in the order you want them to be.
for i in range(len(df.columns)):
    mybox = ax.artists[i]           # Select which box you want to change 

    #mybox.set_facecolor(colors[i])     # Change the appearance of that box
    mybox.set_edgecolor(colors[i])
    
    plt.show()
```

```python
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(14)
```
